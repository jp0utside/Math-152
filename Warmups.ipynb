```python
def boing (y_start, v_start, dt):
  y = y_start
  v = v_start
  for t in range (1000):
    y_new = y + v * dt
    v_new = v - y_new * dt
    v, y = v_new, y_new
    print(y, v)
# boing(2, 0, 0.01)
```

# Warmup 3/2
Central limit theorem: As N -> infinity, the distribution SQRT(N) * (means - mu) approaches N (0, sigma^2)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = np.array([1,1,1,1,6])
mu = np.mean(C)
sigma = np.std(C)
print(mu, sigma)
```

```python
N = 50
samp = np.random.choice(C, N)
print(samp)
print(np.mean(samp))

tests = np.random.choice(C, (1000,N))
means = np.mean(tests, axis = 1)
means.shape
print(means[:10])
plt.hist(means)
plt.show()
```

```python
devs = np.sqrt(N) * (means - mu)
plt.hist(devs, density = True)
x = np.linspace(-6, 6, 1000)
y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*(x/sigma)**2)
plt.plot(x, y, color = "redblue")
```
